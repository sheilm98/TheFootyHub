```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mplsoccer.pitch import Pitch
from mplsoccer import Radar, FontManager
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('fbrefPoss82.csv')

df['Player'] = df['Player'].str.split('\\').str[0]
df['Age'] = df['Age'].str.split('-').str[0]
df = df.drop(columns=['Nation'])
df = df.drop(columns=['Born'])
df = df.drop(columns=['Matches'])
df.head()
```

```python
dfclust = df[['Player', 'Pos', 'Live','Succ%','PrgDist','Rec%']]
dfclust = dfclust.dropna()
dfclust.head()
```

```python
X = np.array(dfclust[['Live','Succ%','PrgDist','Rec%']])
kmeans = KMeans(n_clusters = 10, random_state=100)
kmeans.fit(X)
dfclust['cluster'] = kmeans.predict(X)
dfclust.head()
```
