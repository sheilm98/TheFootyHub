```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("premTbl55.csv")
df.head()
```

```python
df['ScoreHome']= df['Score'].str.split('–').str[0]
df['ScoreAway']= df['Score'].str.split('–').str[1]
df.head()
```

```python
df=df.drop(columns=['Score'])
df=df.drop(columns=['Notes'])
df=df.drop(columns=['Match Report'])
df.head()
```

```python
type(df["ScoreAway"])
df
```

```python
df = df.drop(df[df.Wk > 35].index)
df = df.dropna()
df
```

```python
df
```

```python
type(df['ScoreHome'])
```

```python
winners = []
for index,x in df.iterrows():
    if int(x['ScoreHome']) > int(x['ScoreAway']):
        winners.append(x['Home'])
    elif int(x['ScoreHome']) < int(x['ScoreAway']):
        winners.append(x['Away'])
    else:
        winners.append('Draw')
```

```python
df['Winners'] = winners
df
```

```python
utdwins = []
for index,x in df.iterrows():
        if x['Winners'] == 'Manchester Utd':
            utdwins.append(x)
utdwins
dfutdwins = pd.DataFrame(utdwins)
dfutdwins.head()
```
